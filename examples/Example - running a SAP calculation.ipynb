# Example - running a SAP calculation

This notebook runs SAP10 calculation.

This uses the [sap10calcs Python package](https://github.com/stevenkfirth/sap10calcs).

## 1. Setup

```python
import sap10calcs
```

## 2. Load input XML file

```python
tree, sap_report = sap10calcs.parse_xml('Example - creating a complete SAP XML file.xml')
print(tree)
print(sap_report)
```

## 3. Run SAP calculation

```python
result = sap10calcs.calculate(
    input_lxml = sap_report,
    auth_token = None
)
output_dict = result.pop('sap_calculation_output_dict')
result
```

## 4. View calcuation results

```python
for k, v in output_dict.items():
    print(f'{k}: {v}')
```
